```python
#using power transform for skewness removal:

from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')

pd.DataFrame(x).skew()
```

```python
#scaling the input data

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#as the response variable is of continuous type it is a regression problem

lr=LinearRegression()
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.20)
    lr.fit(x_train,y_train)
    predtrain=lr.predict(x_train)
    predtest=lr.predict(x_test)
    if round(r2_score(predtrain,y_train)*100,2)==round(r2_score(predtest,y_test)*100,2):
        print('At random state',i,'the model performs very well')
        print('At random state',i,':')
        print('Training accuracy score is',r2_score(predtrain,y_train)*100)
        print('Testing accuracy score is',r2_score(predtest,y_test)*100,'\n')
```

```python
#so using state as 71 only:

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=71,test_size=.20)
lr.fit(x_train,y_train)
predtest=lr.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score

for j in range(2,11):
    lss=cross_val_score(lr,x,y,cv=j)
    lsc=lss.mean()
    print('At cv=',j)
    print('Cross Validation Score is:',lsc*100)
    print('Accuracy score is:',r2_score(predtest,y_test)*100,'\n')
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=predtest,color='r')
plt.plot(predtest,y_test,color='b')
plt.xlabel('Actual Happiness Score',fontsize=14)
plt.ylabel('Predicted Happiness Score',fontsize=14)
plt.title('Linear Regression',fontsize=18)
plt.show()
```

# Regularization:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import Lasso

parameters={'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
ls=Lasso(alpha=.001,random_state=0)
ls.fit(x_train,y_train)
lspred=ls.predict(x_test)
lss=r2_score(y_test,lspred)
lss*100
```

```python
lcross=cross_val_score(ls,x,y,cv=10)
lfinal=lcross.mean()
```

```python
lfinal
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(predtest,y_test))
```

# Ensemble technique:

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

parameters={'criterion':['mse','mae'],'max_features':['auto','sqrt','log2']}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mae',max_features='auto')
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)

rfscore=r2_score(y_test,predrf)

print('r2_score:',rfscore*100)

rfcscore=cross_val_score(rf,x,y,cv=5)
rfc=rfcscore.mean()
print('Cross Validation Score:',rfc*100)
```
